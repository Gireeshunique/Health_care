## DATA Cleaning and preprocessing process


```python
df
```

```python
x = df.drop(columns =['PatientId','ScheduledDay','AppointmentDay','AppointmentID'])
```

```python
x
```

```python
x = pd.get_dummies(x, columns =['Gender','Neighbourhood'], drop_first= True)
```

```python
x
```

```python
numerical_columns = x.select_dtypes(include=['int64','float64']).columns
```

```python
numerical_columns
```

```python
from sklearn.preprocessing import StandardScaler


scaler = StandardScaler()
x[numerical_columns]=scaler.fit_transform(x[numerical_columns])
scaler
```

## Data models

```python
from sklearn.model_selection import train_test_split
```

```python
# traget variable is 'Showed_up'
y = df['Showed_up']
x_train,x_test, y_train, y_test = train_test_split(x,y , test_size=0.2, random_state= 42)
```

## radom forest classifier

```python
from sklearn.ensemble import RandomForestClassifier

# intialize the model


model = RandomForestClassifier(random_state = 42)

# train model

model.fit(x_train, y_train)
```

```python
from sklearn.metrics import accuracy_score, classification_report,confusion_matrix
```

```python
# predict the test data

y_pred = model.predict(x_test)

# evaluate  the model
accuracy = accuracy_score(y_test,y_pred)

print(f"Accuracy :{accuracy:.2f}")
# more thwe evaluate the model
print(classification_report(y_test,y_pred))
print(confusion_matrix(y_test,y_pred))
```
